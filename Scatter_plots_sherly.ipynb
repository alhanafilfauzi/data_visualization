# Scatter plots

Untuk membuat scatter plot sederhana, kita menggunakan perintah sns.scatterplot dan menentukan nilai untuk:
sumbu x horizontal (x = insurance_data['bmi']), dan
sumbu y vertikal (y = insurance_datai['biaya']).

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Scatterplot di atas menunjukkan bahwa indeks massa tubuh (BMI) dan biaya asuransi berkorelasi positif, di mana nasabah dengan BMI yang lebih tinggi biasanya juga cenderung membayar lebih banyak biaya asuransi. (Pola ini masuk akal, karena BMI yang tinggi biasanya dikaitkan dengan risiko penyakit kronis yang lebih tinggi).

Untuk mengecek kembali kekuatan hubungan ini, Anda mungkin ingin menambahkan garis regresi, atau garis yang paling sesuai dengan data. Kita melakukan ini dengan mengubah perintah menjadi sns.regplot.




sns.regplot: Ini adalah fungsi dari library Seaborn yang digunakan untuk membuat plot regresi. Plot regresi menunjukkan hubungan antara dua variabel dengan menggunakan garis regresi.

x=insurance_data['bmi']: Ini adalah argumen pertama dari fungsi sns.regplot. Ini mengatur sumbu-x dari plot, yang dalam kasus ini adalah data BMI dari DataFrame insurance_data.

y=insurance_data['charges']: Ini adalah argumen kedua dari fungsi sns.regplot. Ini mengatur sumbu-y dari plot, yang dalam kasus ini adalah data biaya dari DataFrame insurance_data.

Jadi, keseluruhan kode sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges']) menghasilkan plot regresi dengan BMI sebagai sumbu-x dan biaya sebagai sumbu-y, menggunakan data yang ada dalam DataFrame insurance_data.







# Color-coded scatter plots

Kita bisa menggunakan scatter plot untuk menampilkan hubungan antara (bukan dua, tapi...) tiga variabel! Salah satu cara untuk melakukannya adalah dengan memberi kode warna pada titik-titiknya.

Misalnya, untuk memahami bagaimana merokok memengaruhi hubungan antara BMI dan biaya asuransi, kita bisa memberi kode warna pada titik-titik dengan 'smoker', dan memplot dua kolom lainnya ('bmi', 'charger') pada sumbu.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Plot sebaran ini menunjukkan bahwa meskipun bukan perokok cenderung membayar sedikit lebih mahal dengan meningkatnya BMI, perokok membayar JAUH lebih mahal.

Untuk lebih menekankan fakta ini, kita dapat menggunakan perintah sns.lmplot untuk menambahkan dua garis regresi, sesuai dengan perokok dan bukan perokok. (Anda akan melihat bahwa garis regresi untuk perokok memiliki kemiringan yang jauh lebih curam, dibandingkan dengan garis untuk bukan perokok!)

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Perintah sns.lmplot di atas bekerja sedikit berbeda dengan perintah-perintah yang telah Anda pelajari sejauh ini:

Alih-alih mengatur x=insurance_data['bmi'] untuk memilih kolom 'bmi' di insurance_data, kita mengatur x="bmi" untuk menentukan nama kolom saja.
Demikian pula, y = "charges" dan hue = "smoker" juga berisi nama kolom.
Kita menentukan dataset dengan data=insurance_data.
Terakhir, ada satu plot lagi yang akan Anda pelajari, yang mungkin terlihat sedikit berbeda dari cara Anda biasa melihat scatter plot. Biasanya, kita menggunakan scatter plot untuk menyoroti hubungan antara dua variabel kontinu (seperti "bmi" dan "charges"). Namun, kita dapat mengadaptasi desain scatter plot untuk menampilkan variabel kategorikal (seperti "smoker") pada salah satu sumbu utama. Kita akan menyebut jenis plot ini sebagai plot pencar kategorikal, dan kita membuatnya dengan perintah sns.swarmplot.

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

```python
Di antara hal-hal lainnya, plot ini menunjukkan kepada kita bahwa:

rata-rata, pelanggan yang tidak merokok dikenakan biaya lebih rendah daripada pelanggan yang merokok, dan
pelanggan yang membayar paling banyak adalah perokok; sedangkan pelanggan yang membayar paling sedikit adalah non-perokok.
```
